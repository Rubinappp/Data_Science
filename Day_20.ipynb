## reshape function

__numpy.reshape(a, newshape, order='C')__<br>
 a        =Array to be reshaped.<br>
 newshape =int or tuple of ints<br>
 order    ={‘C’, ‘F’, ‘A’}, optional
 
C-order ('C'):Elements are stored row by row. The last index (columns) changes fastest..<br>
Fortran-order ('F'): Elements are stored column by column. The first index (rows) changes fastest.Here,Fill column first<br>
Automatic ('A'): Preserves the input array's memory order.



* This function is very useful when we need to change the dimensions of our data for various calculations or visualizations.
* The number of elements in the original array must match the total number of elements in the reshaped array

```python
np.reshape(a,(6,2),"F")
```

```python
np.reshape(a,(6,2),"C")
```

```python
np.reshape(a,(6,2),"A")
```

2. __np.random.randn(no.of random numbers)__


The np.random.randn function generates numbers from a standard normal distribution, which is a bell-shaped curve symmetric around zero. In this distribution:

The mean (average) value is 0.
The standard deviation is 1.

The values can range from negative to positive, with about 68% of the values falling within one standard deviation of the mean (i.e., between -1 and 1), about 95% within two standard deviations (between -2 and 2), and so on.

## `concatenate()` function 

`concatenate((a1, a2, ...), axis=0)`

- Join a sequence of arrays along an existing axis.

- a1, a2, ... : sequence of array_like
  - The arrays must have the same shape, except in the dimension corresponding to `axis` (the first, by default).
- axis : int, optional
  - The axis along which the arrays will be joined.
  - If axis is None, arrays are flattened before use.  Default is 0.
  

axis = 0 → join vertically -> diff columns are allowed

(put one array below another)

axis = 1 → join horizontally -> diff rows are allowed

(put one array beside another)

```python
arr1=np.arange(2,20).reshape(6,3)
```

```python
arr2=np.arange(1,13).reshape(6,2)
```

```python
np.concatenate([arr1,arr2],axis=0)
```

```python
np.concatenate([arr1,arr2],axis=1)
```
