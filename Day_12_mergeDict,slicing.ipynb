##### merge two dictionary without using function directly

```python
dict1={"apple":"10kg","orange":"24kg"}
dict2={"mango":"5kg","watermelon":"20kg"}

dict3=list(dict2.items())+list(dict1.items())
for i,j in dict3:
    dict1[i]=j

print(dict1)


```

```python
dict1={"apple":"10kg","orange":"24kg"}
dict2={"mango":"5kg","watermelon":"20kg"}



for i,j in dict1.items():
    dict2[i]=j

print(dict2)
```
