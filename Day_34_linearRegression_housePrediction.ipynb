# Transformation:

### 1. Min–Max Normalization (0 to 1 scaling)

Also called MinMaxScaler.

Compresses data into [0, 1]

Useful for methods sensitive to absolute scale (KNN, neural networks)

### 2. Z-Score Normalization (Standardization)

Also called StandardScaler.



After this:

- Mean becomes 0

- Standard deviation becomes 1

For MOST classical ML algorithms → Use Standardization (Z-Score)

1. Linear Regression / Logistic Regression

2. SVM

3. SGD Regressor / Classifier

4. K-Means

5. PCA

6. Lasso / Ridge

7. Naive Bayes

8. Gaussian Mixture Models

#### Why?

#### Standardization makes features:

- mean = 0

- std = 1

- symmetric around zero

Normal distribution data → z-score → standard normal (bell-shaped)

Skewed data → z-score → still skewed, just shifted and scaled

→ This helps gradient descent converge faster

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train_scaled,Y_train)
```

```python
y_pred=model.predict(X_test_scaled)
```

```python
Y_test,y_pred
```

```python
model.coef_ # slopes of features w1,w2,...
```

```python
model.intercept_
```

```python
model.score(X_test_scaled, Y_test) #r2 score
```
