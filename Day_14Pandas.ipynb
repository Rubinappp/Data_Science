## key/value object as series in dictionary


#### A Series in Pandas is essentially a one-dimensional array-like object with an associated array of data labels, called its index. When you create a Series from a dictionary, each key-value pair in the dictionary becomes an index-value pair in the Series. This is straightforward because a Series is fundamentally a single column of data.


```python
import pandas as pd
calories={"day1":420,"day2":380,"day3":450}
nums=pd.Series(calories)
```

```python
nums
```

```python
import pandas as pd
calories={"day1":420,"day2":380,"day3":450}
nums=pd.Series(calories,index=["day1","day2"])
```

```python
nums
```

#### When you provide a dictionary to create a DataFrame, Pandas expects the dictionary keys to represent columns and the values to be array-like structures (lists, Series, arrays, etc.) that define the rows of each column. This is why you need to provide the data as a list of dictionaries or a dictionary of lists.


#### suppose i have calories={"day1":420,"day2":380,"day3":450} then pandas become confuse wether to treat it as 1 row and 3 coumns or 3 row and 1 column. It doesnt know how to treat scalar which is why either we put index manually or make it list.



```python
df=pd.DataFrame([calories])
df
```

```python
df=pd.DataFrame(calories,index=["0"])
df
```
