##### find sum of all even numbers from 1 to 21

```python
list1=list(range(1,22))
sum=0
odd_sum=0
for i in list1:
    if(i%2==0):
        sum=sum+i
    else:
        odd_sum=odd_sum+i
print("even number sum",sum)
print("odd number sum",odd_sum)
```

```python
i=0
sum=0
odd_sum=0
while(len(list1)>i):
    if(list1[i]%2==0):
        sum=sum+list1[i]
    else:
        odd_sum=odd_sum+list1[i]
    i=i+1;
print("even number sum",sum)
print("odd number sum",odd_sum)  
```
