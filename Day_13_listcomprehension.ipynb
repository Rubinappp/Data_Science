#### To find odd and even numbers in list

```python
list1=[1,2,3,4,5,6,12]
even=[]
odd=[]
for i in list1:
    if(i%2==0):
        even.append(i)
    else:
        odd.append(i)
print(even)
print(odd)
```

## list comprehension
List comprehension is a shorter and more readable way to create lists in Python.

Basic Syntax

[expression for item in iterable if condition]


```python
list1=[1,2,3,4,5,6,12]
```

```python
[i for i in list1 if(i%2==0)]
```

```python
a=[i for i in list1 if(i%2==0)]
print(a)
```

### Enumerate
enumerate() is a built-in Python function that lets you loop over a list (or any iterable) and keep track of the index of each item at the same time.

enumerate(iterable, start=0)


```python
list1=[1,2,3,4,5]
for idx,num in enumerate(list1):
    print(idx,num)
```

#### Input: ["Ram","shyam","hari"]  Expected Output: [(0, 'Ram'), (1, 'shyam'), (2, 'hari')]

```python
namelist=["Ram","shyam","hari"]
list1=[]
```

```python
for idx,name in enumerate(namelist):
    list1.append((idx,name))
    
```

```python
list1
```

using comprehension

```python
[(idx,name) for idx,name in enumerate(namelist)]
```

#### Write a program to print words greater than 4 and append inside list

without using list comprehension

```python
sentence="Hello guys how are you doing? I am doing well."
sentence_1=sentence.split()
sentence_2=[]
print(sentence_1)
```

```python
for i in sentence_1:
    if(len(i)>4):
        sentence_2.append(i)
```

```python
print(sentence_2)
```

using list comprehension

```python
[ i for i in sentence_1 if(len(i)>4)]
```
