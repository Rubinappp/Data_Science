##### create a function called filter_sum that takes a list of numbers and return the sum of only the numbers that are greater than 10.

```python
def filter_sum(list1):
    sum=0
    for i in list1:
        if i>10:
            sum=sum+i
    return sum


list1=[1,10,20,30]
sum=filter_sum(list1)
print(f"the sum of the numbers that are greater than 10 is {sum}")
```

##### Create a function that takes a sentence as input and returns the occurrence of words. (Hint: split())

```python
def count_word(sentence):
    splitted = sentence.split(" ")
    # new list which has lowercase of all words 
    new = []
    for word in splitted:
        new.append(word.lower())

    #counter dict to save the counts of words
    counter = {}

    # created set for only having the unique words which would help to loop through a word only once
    splitted_unique = list(set(new))
    print(splitted_unique)
    
    for word in splitted_unique:
        temp_count = new.count(word)
        counter[word] = temp_count
    return counter

count_word("This is a very beautiful city. this city")
```

##### write a function to return odd and even numbers

```python
def odd_even(num : list, reteven : bool = True ):
    list1=[]
    list2=[]
    if reteven:
        for i in num:
            if(i%2==0):
                list1.append(i)
        return list1
    else:
        for i in num:
            if(i%2!=0):
                list2.append(i)
        return list2
num=[1,2,3,4,5,6,7]
result=odd_even(num)
print(result)
result1=odd_even(num,reteven=False)
print(result1)
```
