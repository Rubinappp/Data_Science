##### Create a class House with the following:

A constructor that takes the number of rooms, number of floors, color, and an optional garden argument (default is False).

A method detail() that prints the number of rooms, color, and floors of the house.

A method has_garden() that prints whether the house has a garden or not.

```python
class House:
    def __init__(self,rooms,floor,color,garden=False):
        self.rooms=rooms
        self.floor=floor
        self.color=color
        self.garden=garden
    def detail(self):
        print(f"The house has {self.rooms} rooms.The color of house is {self.color}.It has {self.floor} floors")
    def has_garden(self):
        if self.garden:
            print("the house has garden")
        else:
            print("the house doesnot have garden")
    
house1=House(6,3,"white")
house1.detail()
house1.has_garden()
    
house2=House(6,3,"blue",True)
house2.detail()
house2.has_garden()
```

##### Implement a simple banking system using a class in Python that allows a customer to deposit money, withdraw money, and check their balance.

```python
class Bank:
    def __init__(self,customer_name,phone_no,balance):
        self.customer_name=customer_name
        self.phone_no=phone_no
        self.balance=balance
    def deposit(self):
        amount=int(input("enter the amount to deposit:"))
        self.balance+=amount
        
    def withdraw(self):
        amount=int(input("enter the amount to withdraw:"))
        if(amount<self.balance):
            self.balance-=amount
           
        else:
            print("error!Exceeded limit")
        
    def show_balance(self):
        print(f"Your balance is: {self.balance}")
```

```python
customer1=Bank("Rubina",9873628,40000)
customer1.deposit()
customer1.show_balance()
```

```python
customer1.withdraw()
customer1.show_balance()
```
