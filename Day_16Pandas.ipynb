```python
## Task 9 : What is the average price of cars?
```

```python
carData['Price']
```

```python
a=""
sum=0
count=0
for i in carData['Price']:
    count=count+1
    a=i[1:]
    k,j= a.split(',')
    final=k+j
    sum=sum+float(final)
    
average=sum/len(carData['Price']) # sum/count
print(average)
 
```

```python
# using list comprehension
```

```python
def cleanPrice(price):
    price = price[1:]
    total = ""
    for p in price.split(","):
       
        total += p
      
   
    return float(total)

#list comprehension
carData["Price(float)"] = [cleanPrice(price) for price in carData['Price']]   

carData["Price(float)"].mean()
```

```python
#Another way
```

```python
def fun(j):
    price=j[1:]
    j,k= price.split(',')
    return (float(j+k))
    
l=pd.Series([fun(i) for i in carData['Price']])
l.mean()
```
