Correlation Analysis

It is statistical method to measure the relationship between two variables

we generally want to know two things in correlation analysis

1. How strong the correlation is.
2. In which direction the correlation is going.

we can know both by finding correlation coefficient value that ranges from -1 to 1

Pearson's correlation is one of the method to calculate corelation coefficient.It also ranges from -1 to 1

Negative correlation has downward trend when x increase y will decrease.

eg of negative correlation:
    mistakes vs practice hour.
    if practice hour (x) increases then (y) mistakes gets decreases.
 

For using pearson coefficient some criteria must be met

1. Linearity:change is one variable associated with a change in proportional another variable.if relationship is non linear could be misleading and may not capture true real relationship.

2. Bivarate Normally:
both variable should be normally distributed and the joint distribution of the two variable should also be normal

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# it was historically required to make plots appear in cells.
%matplotlib inline 



# Generate data
x = np.random.rand(100)  # 100 random numbers from 0 and 1
y = -2.5 * x + np.random.normal(0, 0.1, 100)  # y is linearly dependent on x with some noise
# to make noise realistic so that value doesnt get far away from mean we used normal here not rand.
# mean=0 ,sd=0.1,size=100

# Put data into a DataFrame
data = pd.DataFrame({'X': x, 
                     'Y': 2.5 * np.random.rand(100) + np.random.normal(0, 0.1, 100),
                     'z':-1.5*np.random.rand(100) + np.random.normal(0, 0.1, 100),
                      'w':np.random.normal(0,1,100),
                      'v':y})
plt.figure(figsize=(4,3))
# calculate correlation matrix
corr=data.corr()
#Create a heatmap
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f',linecolor='black',linewidth=1)
plt.title('Heatmap of Correlation Matrix')

plt.tight_layout()

plt.show()
```

```python
plt.scatter(data['X'], data['z'])
```

Parameters:

annot=True: This parameter adds numeric annotations to each cell of the heatmap, displaying the correlation coefficients.

cmap='coolwarm': This specifies the color map to be used for the heatmap. 'coolwarm' is a popular choice where negative correlations are displayed in cool tones (blue) and positive correlations in warm tones (red).

fmt='.2f': This controls the format of the annotations. '.2f' specifies that annotations should be displayed as floating-point numbers with two decimal places.

linecolor='black': This sets the color of the lines that divide each cell in the heatmap.

linewidth=1: This determines the width of the lines that divide each cell in the heatmap.
