```python
import pandas as pd
```

```python
# Creating a sample dataset
data = {
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Eve', 'Charlie', 'David', 'Frank', 'Grace', 'Helen'],
    'Subject': ['Math', 'Science', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science'],
    'Score': [85, 78, 92, 88, 75, 80, 98, 79, 88, 90]
}

df = pd.DataFrame(data)
```

```python
df
```

### Aggregation = taking a bunch of values and summarizing them into a single value.
'mean','sum','max','min','count'

## To find the mean() score based on particular subject
## We use groupby() method of pandas

```python
average_score=df.groupby('Subject')['Score'].mean()
```

```python
average_score
```

```python
average_score[0] # this wont be used in future
```

## Finding max and min of scores

```python
max=df.groupby('Subject')['Score'].max()
```

```python
max
```

```python
min=df.groupby('Subject')['Score'].min()
```

```python
min
```

## Count scores in particular subject

```python
count=df.groupby('Subject')['Score'].count()
```

```python
count
```

## Find total scores of each subject

```python
sum=df.groupby('Subject')['Score'].sum()
```

```python
sum
```

## Performing multiple Aggregations

```python
agg_results=df.groupby('Subject')['Score'].agg(['mean','sum','max','min','count'])
```

```python
agg_results
```

```python
df.groupby("Subject").agg({'Score':'mean'})
```

```python
df.groupby('Subject')['Score'].agg(
                          average='mean',
                          add='sum')
```

## Performing Different aggregation for different column

```python
mydata = {"name": ["Ram", "Shyam", "Ram"],
          "a":[1, 2, 3],
          "b" : [4, 5, 6],
          "c":[7, 8, 9]}
tempdf = pd.DataFrame(mydata)
tempdf
```

```python
tempdf.groupby("name").agg({'a':'mean','b':'max'})
```
