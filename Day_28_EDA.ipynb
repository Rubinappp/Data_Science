# Important concepts in descriptive analysis

### a. Measures of Central Tendency

#### 1.Mean :Arithmetic Average of a dataset.
- Calculations:Sum of all measurements divided by the number of observations.
- Use Case:Best used when data distribution is symmetrical and without outliers.

#### 2.Median:Middle value in a dataset when it is ordered from smallest to largest
- Calculations:If the number of observation is odd,it is the middle element;if even,it is the average of the two middle numbers.
- Use Case:More robust than the mean in skewed distributions.

#### 3.Mode:The value that appears most frequently in a data set.
- Use Case:Useful in analyzing categorical data.


```python
import numpy as np
a=[1,3,5,7,9]
np.mean(a)
```

```python
np.median(a)
```

Here both mean and median are same. 
But mean will get heavily affected if there are outliers.

- Mean breaks when data has outliers

- Median stays stable and trustworthy

Let us take example:

```python
import numpy as np
a=[1,3,5,7,900]
np.mean(a)
```

```python
np.median(a)
```

- Mean is affected by extreme values (outliers), which can distort the representation of the central tendency if the dataset contains outliers.
- Median remains unchanged and better represents the central location of the dataset in the presence of outliers.

This illustrates why the median is often preferred over the mean when dealing with skewed distributions or datasets with outliers.







### b. Measures of Dispersion

Statistical technique used to describe spread and variability within a dataset

They provide insights on distribution of datasets.How similar or diverse the values are.

#### 1.Range:
- Range=Maxvalue-Minvalue

- Gives quick sense of spread of data but can be highly sensitive to outliers.


- If the range is small, the max value is close to the min → data is “compressed” or tightly packed.

- If the range is large, the max value is far from the min → data is more “stretched out”.

What happens if there’s an outlier?

Example:

Data: [5, 6, 5, 6, 5, 100]


Min = 5

Max = 100

Range = 95 → looks huge (tells data is stretched)

But most of your data (5,6,5,6,5) is tightly clustered around 5–6.

Range suggests the data is “stretched out”,but that’s not true for most of the data because only one value (100) is far away.

So range overestimates the actual spread of most values
#### 2.Interquartle Range(IQR):
- Measure of Range within which central 50% of values fall.
- IQR=Q3-Q1
- More robust than range as it is not affected by outliers or extreme values.
- IQR tells you how “stretched” the central portion of our data is, ignoring extremes.

Use of IQR 
- for calculation of outliers.
- In box plot IQR is represnted by box's length

```python
import numpy as np

# Sample data
data = np.array([9, 2, 5, 4, 12, 7, 8, 11, 9, 3, 7, 4, 12, 5, 4, 10, 9, 6, 9, 4])
# Calculating measures of dispersion

#Peak to peak (max-min) function or range
range_data=np.ptp(data)
iqr_data=np.percentile(data,75)-np.percentile(data,25)

variance_data=np.var(data,ddof=1)
st_dev_data=np.sqrt(variance_data)

print(range_data,",",iqr_data,",",variance_data,",",st_dev_data)
```

ddof=1 gives slightly larger variance for the same data, because dividing by n-1 instead of n corrects for the fact that a sample usually underestimates the population variance.(means gives smaller value than it should)

## c. Shape of distribution
Normal Distribution

The normal distribution is also known as a Gaussian distribution or probability bell curve. It is symmetric about the mean and indicates that values near the mean occur more frequently than the values that are farther away from the mean.50%-50% split This doesn’t mean data is evenly spread in each half most values are concentrated or clustered around the mean.

In a perfect normal distribution:

mean = median = mode

Example: some people do good on exam and some people do worst but majority scores average.


Charactersitics:
1. Symmetry:Distribution is perfectly symmetrical around its mean. Right side of the center is mirror image of left side.
2. Mean,Median and Mode : In perfectly normal distribution they are all equal and lie at center of distribution.
3. Skewness:A normal distribution has a skewness of zero (perfect symmetry)

on the basis of value of mean value the curve shift left and right.

larger the standard  deviaion more spread the curve is .
Standard deviation (σ) measures spread.

Large σ → values farther from mean → lower density at center-> curve is wider






smaller the sd less spread.

Small σ → most values are close to mean → high density-> curve is narrower and tall





normal distrubution is density curve whole total area needs to be 1 or 100% so change in width must be compensated for height of curve.



##### Example1.

```python
import numpy as np
import matplotlib.pyplot as plt

#Generate normally distributed data
np.random.seed(0)
data_normal=np.random.normal(loc=50,scale=10,size=1000)

#Create figure and subplots
fig,axs=plt.subplots(2,1,figsize=(20,8))
axs[0].hist(data_normal,bins=30,alpha=0.7,color="blue") # alpha =transparency (opacity) 0.0 means full transparent 
axs[0].set_title("Histogram of normally Distributed Data")
axs[1].boxplot(data_normal,vert=False)
axs[1].set_title("Boxplot of Normally Distributed Data")
plt.show()
```

loc=50 → mean = 50

scale=10 → standard deviation = 10

size=1000 → total 1000 samples

##### Example2:Outlier

outlier = which is different from majority of values.

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data_with_outliers=np.random.normal(loc=50,scale=10,size=1000)
outliers=data_with_outliers[:10]+100
data_with_outliers[:10]=outliers

fig,axs=plt.subplots(2,1,figsize=(20,8))
axs[0].hist(data_with_outliers,bins=30,alpha=0.7,color="green")#bins = number of rectangles (bars)
axs[0].set_title("Histogram of Data with Outliers in One tail")
axs[1].boxplot(data_with_outliers,vert=False)
axs[1].set_title("Boxplot Showing outliers in One Tail")
```

```python
import seaborn as sns

sns.histplot(data_with_outliers,kde=True)
plt.show()
```

It is right skewed distribution 

```python
from scipy import stats
stats.mode(data_with_outliers)
```

```python
np.median(data_with_outliers)
```

```python
np.mean(data_with_outliers)
```

outliers in a dataset as similar to "outcasts" in a social context. : Outliers are data points that are significantly different from the majority of the data points. They lie far from the central tendency (mean, median, or mode) of the dataset.

Identification: Outliers are identified using statistical methods such as z-scores, interquartile range (IQR), or visual tools like box plots.

In data, outliers can be mistakenly removed when they might hold valuable information. In society, outcasts can be unfairly judged or excluded without understanding their potential contributions.
 
 Outliers are crucial in data analysis for understanding variability and identifying potential anomalies.
