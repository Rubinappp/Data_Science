## Subplots

- Subplots allow us to place multiple plots in a grid layout within a single figure.
- This is useful for comparing different datasets or different views of the same data side by side.

```python
import matplotlib.pyplot as plt
import numpy as np

day_count = np.arange(1, 91)
weight = np.linspace(85, 60, 90)
print(weight)
print(day_count)
```

```python
figure,(ax1,ax2)=plt.subplots(1,2,figsize=(20,10)) # 1*2=2 (1 row 2 column)subplots(axis)
figure.suptitle("Different Plots for Weight Loss")

ax1.plot(day_count,weight)

ax1.set_title("Line plot of weight loss")
ax1.set_xlabel("Day")
ax1.set_ylabel("Weight")

ax2.scatter(day_count,weight)
ax2.set_title("Scatter plot of weight loss")
ax2.set_xlabel("Day")
ax2.set_ylabel("Weight")
plt.show()
```

```python
x=np.arange(1,10)
y=x
y2=np.power(x,2)
```

```python
f,(ax1,ax2)=plt.subplots(2,1,sharex=True,figsize=(20,5)) # sharex=True will share the x axis of both subplot
f.suptitle("sharing y axis")

ax1.plot(x,y)
ax2.plot(x,y2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(10)
y1 = [2 * xi for xi in x]
y2 = [xi**2 for xi in x]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 5))
fig.suptitle('Vertically Stacked Subplots')

ax1.plot(x, y1)
ax1.set_title('Subplot 1')

ax2.plot(x, y2)
ax2.set_title('Subplot 2')

plt.show()
```

```python
import pandas as pd
```

```python
weatherdata=pd.read_csv("WeatherData.csv",index_col=0,parse_dates=True)
```

index_col=0: This parameter sets the first column (0-based index) of the CSV file as the index of the DataFrame. 

parse_dates=True: date lai date anusar liga vanna ko lagi 

```python
weatherdata.head()
```

```python
date=weatherdata.index
print(date)
date=list(date)
print(date)
temperature=weatherdata["Temperature"]
Rainfall=weatherdata["Rainfall"]
```

```python
figure,(axs1,axs2)=plt.subplots(2,1,figsize=(20,8))

axs1.plot(date,temperature,label="Daily Temperature",color='red')
axs1.set_title("Daily Temperature over a year")
axs1.set_ylabel("Temperature(C)")
axs1.legend()

axs2.bar(date,Rainfall,label='Daily Rainfall',color='blue',alpha=0.5)
axs2.set_title("Daily Rainfall over a year")
axs2.set_ylabel("Rainfall(mm)")
axs2.legend()
plt.tight_layout()
plt.show()
```
