## What is NumPy?
stands for numerical python, is a  fundamental package for scientific computing in python.It typically hold numeric data. 

While NumPy is optimized for numbers, it can also store non-numeric data like strings, booleans, or dates — as long as the data is still homogeneous.

If we need to store heterogeneous data (e.g., a mix of numbers, strings, lists), you must explicitly set the array's dtype to object. This allows elements to be of different Python types, but comes at the cost of performance and loss of vectorization.

1.provides high performance multidimensional array object and tools for working with these array.

2.It is a linear algebra library for data science with python.

3.Bindings to C libraries, making operations significantly faster than native Python lists.(loopings are handled internally in C in numpy)

4.array structure comes in two forms:

vectors(1-d array) and matrices(2-d array) (and also support higher dimensions)


5.Vectroized operation (element wise operation)

6.contiguous memory allocation

To calculate address for a 1D Numpy array:

address = base_address + index * itemsize

## Why NumPy is Faster?

### Data Structure:
NumPy arrays are homogenous and consist of fixed type(typically numerical data)wchich allows for more efficient storage in memory .In contrast python lists are 
heterogenous which mean they can contain element of different types and each element is stored as a seperate object in memory.

### Memory Layout:
NumPy uses contiguous memory which can be efficiently cached and has benefits from vectorized operations.Python list however are arrays of pointers to objects ,even
when all of them are of the same type

### Built in Operations:
NumPy operations are implemented in C which mean they are compiled and highly optimized for performance.Python's sum() function,while versatile(can handle different types), is implemented in python and iterates in bytecode  which is inherently slower than operations in compiled C used by NumPy.

### Vectorized Operations:
vectorization in NumPy means performing operations element-wise without using explicit Python loops.
Compare the use of a Python loop vs. vectorized operations for multiplying elements by2

```python
t=[1,2,3]
t*2
```

```python
temp=[]
for num in t:
    temp.append(num*2)
temp    
```

```python
z=np.array([1,2,3])
z*2
```

```python
a=np.array(range(1,13))
print(a)
```

```python
%timeit [i*2 for i in a]
# underhood there is for loop which ranges 100000 times under that 
#we have our code .so 100000 times the code run and average of the result is calculated to get the accurate result
%timeit a*2
```

## Basic Functions:

__np.sum()__

```python
y=np.array([[1,2],[3,4]])
print(y)
```

```python
print(np.sum(y))
```

```python
print(np.sum(y,axis=0)) # sum of each column
```

axis 0 is row axis.. so we add all rows together column by column (1 row is getting added to another row)

```python
np.sum(y,axis=1) # sum of each row
```

axis 1 is the column axis. so 1 column is added to another column
