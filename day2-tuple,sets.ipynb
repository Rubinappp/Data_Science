Tuples cannot be modified after creation, but you can create a new tuple by concatenating existing tuples.

```python

my_tuple = (1, 2, 3)
new_tuple = my_tuple + (4,)  # Create a new tuple by concatenation
print(new_tuple)
```

### 2. Sets

##### holds unique items
##### no key value pair like dictionary

```python
numbers={1,1,2,2,3,4,5,6,7,10,10} 

print(numbers)
len(numbers)
#repititive items still prints  unique items
```
