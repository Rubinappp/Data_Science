### 3.Formatting with placeholders (%s,%r,%d format specifier )

##### %s:  to inject strings into your print statements. The modulo % is referred to as a "string formatting operator".



```python
print("I'm very %s today." %'happy')
```

```python
print("The %s seems to be %s." %('weather','perfect'))
```

##### %r and repr(): deliver the string representation of the object, including quotation marks and any escape characters.

```python
print('He said his name was %r.' %'Fred')
```

##### \t inserts a tab into a string.

```python
print("the cat is sitting on %s"%"the\t mat")
print("the cat is sitting on %r"%"the\t mat")
```

##### %d : converts numbers to integers (removing the decimal part), without rounding.

```python
print("I wrote %s programs today." %3.755)
print("I wrote %d programs today." %3.755)
```

### dictionary.setdefault(key, default_value)
 
 It checks:

- If key exists:

→ Returns the current value (does not change it)

- If key doesn’t exist:

→ Sets dict[key] = default_value ( add new key value pair)

```python
rollnums={1:"ram",2:"shyam" ,3:"hari",}
rollnums.setdefault("ok")
print(rollnums)

rollnums.setdefault(4,"pk")
print(rollnums)
```

### Hashmap reffered to as dictionary in python
- Its a data structure.

Benefits:
  
- customize key for readability.

- allow for search in O(1) which is faster than array/linkedlist O(n).

A hashmap or dictionary:

- Stores data in key-value pairs.

- Each dict in Python has its own hash table, stored as part of its internal structure in memory.

- Uses a hash function to compute an index for storing the value. Hash function take key and gives hash value(index) where value is stored.

- Ensures each key is unique.

- If a key can change, its hash value can also change, and Python wouldn't be able to find it again in the dictionary. That’s why mutable objects like sets are not allowed as keys.

```python
dict2={{3,4}:"poudel"} # using mutable set
```

#### Note:
#### unhashable = Can't generate stable index because its mutable so cannot be used as key.

```python
dict1={(1,2):"poudel"} # using immutable tuple
dict1
```

#### sets,list dict are unhashable (mutable)
#### tuple is hashable(immutable)
#### tuple can be unhashable if list is present inside it.

- dictionary uses hash value to store.so key must be unique and unhashable( giving stable index)

-  tuple may or  may not use hash value .May not use if it store list or dictionary inside it.

-  sets use hash value to store so doesnt allow any unhashable items inside it.

-  list doesnt use hash value and unhashable at all.
