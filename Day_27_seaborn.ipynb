```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a basic DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
        'Age': [25, 30, 35, 40, 45],
        'Salary': [50000, 60000, 70000, 80000, 90000]}

df = pd.DataFrame(data)

# Print the DataFrame
df
```

```python
# Create scatter plot of Age vs Salary
sns.scatterplot(x='Age',y='Salary',data=df)
# Show the plot 

plt.title('Age vs Salary')
```

```python
# Create scatter plot of Age vs Salary
plt.figure(figsize=(20,9))
sns.set_style(style='whitegrid')
sns.scatterplot(x='Age',y='Salary',data=df,hue='Name',size='Salary',sizes=(50,200))
# Show the plot 
plt.title('Age vs Salary')
plt.show()
```

```python
# Create scatter plot of Age vs Salary
plt.figure(figsize=(20,9))
sns.set_style(style='whitegrid')
sns.scatterplot(x='Age',y='Salary',data=df,hue='Name',legend='full')
# Show the plot 
plt.legend()
plt.title('Age vs Salary')
plt.show()
```

```python
# Create scatter plot of Age vs Salary
sns.scatterplot(x='Age',y='Salary',data=df)
plt.xlabel("x 1")
plt.ylabel("x 2")
# Show the plot 
plt.title('Age vs Salary')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a basic DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
        'Age': [25, 30, 35, 40, 45],
        'Salary': [50000, 60000, 70000, 80000, 90000],
       'Gender' : ["Male", "Male", "Female", "Female", "Male"]}

df = pd.DataFrame(data)

# Print the DataFrame
df
```

```python
plt.figure(figsize=(20,9))
sns.set_style(style='whitegrid')
sns.scatterplot(x='Age',y='Salary',data=df,hue='Gender')
# Show the plot 
plt.title('Age vs Salary')

plt.show()
```
