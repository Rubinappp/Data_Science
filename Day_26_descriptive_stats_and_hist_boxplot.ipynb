###  In January 1985,the space shuttle challenger broke apart shortly after liftoff . The accident was caused by a part that was not designed to fly at the unusually cold temperature at 29 degree F at launch.

```python
import matplotlib.pyplot as plt
```

```python
temperatures = [66,70,69,80,68,67,72,70,70,57,63,70,78,67,53,67,75,70,81,76,79,75,76,58,29]
```

```python
len(temperatures)
```

```python
plt.figure(figsize=(18,5))
plt.hist(temperatures)
plt.xlabel("Temperatures in F")
plt.ylabel("Frequency")
plt.title("Launch Temperatures of the first 25 shuttle missions in F")
plt.show()
```

The default value of the number of bins to be created in a histogram is 10.

However, we can change the size of bins using the parameter bins(also called classes or intervals)

```python
plt.figure(figsize=(20,9))
bins=range(min(temperatures),max(temperatures)+1)
plt.hist(temperatures,bins=bins,edgecolor='black') # edgecolor=border
plt.xticks(bins) # to gives x values on own 
plt.xlabel("Temperatures in F")
plt.ylabel("Frequency")
plt.title("Launch Temperatures of the first 25 shuttle missions in F")
plt.show()
```

## Box Plots(Box and whisker plots)

A  box plots also known as  box-and-whisker plot, is a standardized way of displaying the distribution of a data set based on its five-number summary of data points: 

- minimum
- first quartile [Q1],
- median
- third quartile [Q3] 
- maximum.

##### It provides a visual summary of several important charactersitics of data set such as center ,spread,skewness and the presence of outliers.

```python
import pandas as pd
```

```python
data=pd.read_csv("BikeData.csv",index_col=0)
```

```python
data
```

```python
plt.figure(figsize=(10,3))
plt.boxplot(data["Engine Capacity (cc)"],vert=False,patch_artist=True)
plt.show()
```


1. min = smallest value=0th percentile
2. Q1 = 25th percentile is a single value such that:

    25% of the data values are less than or equal to Q1
    And
    75% of the data values are greater than or equal to Q1


3. median (Q2) = 50th percentile

4. Q3 = 75th percentile

5. max = largest value=100th percentile


Q1 IS the median of the lower half of the data.

Q3 IS the median of the upper half of the data.

- IQR = Q3  Q1 measure spread(how far apart the middle part of the data is)
- Outlier = value outside [Q1  1.5IQR, Q3 + 1.5IQR]
- lower outlier is < Q1-1.5*IQR (Inter quartile range Q3-Q1)

- higher outlier is > Q3+1.5*IQR
