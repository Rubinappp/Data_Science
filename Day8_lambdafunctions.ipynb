##### Higher order functions (HOFs) : 

### map in python
It's a function that takes another function and iterable object as parameter. 

##### When you write:


map(function, iterable)


- It takes the first item from the iterable.

- It calls the function with that single item.

- It return the result immediately.

- Repeats for the next item.

- all results are collected into a list(if wrapped inside list) and printed as a whole.

##### Write a function that calculates the area of a square you have 10 length of side of squre in list

##### Without map function:

```python
def area(list1):
    return(list1**2)
    

list1=[2,5,7,10,15]
list2=[]
for i in list1:
    list2.append(area(i))

print(list2)
```

##### With map function:

```python
def area(list1):
    return(list1**2)
list1=[2,5,7,10,15]
list2=[]
print(map(area,list1))
list(map(area,list1))
```

Here, It reduced use of for or any explicit loop to call that function and append.

##### Find Length of Strings using map function

```python
words=["hello","hi","konichiwa"]
list(map(len,words))
```

##### convert temperature from celsius to fahreinheit

```python
def convert(list1):
    return(9/5)*list1+32
    
list1=[25,30,40,32]
list(map(convert,list1))
```

```python
def odd_even(list1):
    if (list1%2==0):
        return(f"{list1}:even")
    else:
        return(f"{list1}:odd")

list1=[2,3,4,5,6]
print(list(map(odd_even,list1)))
```

## filter 
filter(function,iterable)
##### test if item is accepted or not
##### it returns the original values for which the function returns True.

```python
def is_non_negative(num):
    return num>=0
numbers=[-10,-5,0,5,10]
non_negative_numbers=filter(is_non_negative,numbers)
print(list(non_negative_numbers))
```

```python
def even(num):
    return(num%2==0)
numbers=[2,5,10,20,25]
even_numbers=filter(even,numbers)
print(list(even_numbers))
```

### lambda functions



lambda parameter:return expression

##### anonymous function, often passed as an argument to higher order function

##### cannot have multiple expression must be a single expression.

##### without using lambda function

```python
def area(a):
    return(a**2)
result=area(2)
print(result)
    
```

##### using lambda function

```python
area=lambda num:num**2
result=area(2)
print(result)
```

use a lambda function to square each number in the list.

```python
numbers=[1,2,3,4]
square_num=map(lambda num:num**2,numbers)
print(list(square_num))
```

uses a lambda function to check if the number is even or odd.

```python
even_odd=lambda num:num%2==0
print(even_odd(int(input("enter the number"))))
```

lambda expression to return only the even numbers from the list.

```python
list1=[1,2,3,4,5]
even=filter(lambda num:num%2==0,list1)
print(list(even))
```
