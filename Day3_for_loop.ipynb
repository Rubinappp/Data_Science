#### print [ (12, 45), (13, 56), (34, 67) ] using for loop

```python
a=[12,13,34]
b=[45,56,67]
for i in range(0,3):
    print((a[i],b[i]))
   
    
```

```python
a=[12,13,34]
b=[45,56,67]
for i in a:
    for j in b:
        print(i,j)
    print("we are outside j")
```

```python
a=[12,13,34]
b=[45,56,67]
result=[]
for i in range(0,len(a)):
    result.append((a[i],b[i]))
print(result)
```


###### tupples inside list

```python
a = [('apple', 1, 23), ('banana', 1, 23), ('guava', 10, 23), ('mango', 5, 23)]

for i in a:
    print(i)
```

```python
a = [('apple', 1, 23), ('banana', 1, 23), ('guava', 10, 23), ('mango', 5, 23)]

for i, j, k in a:
    print(i, j, k)
```

```python
a={"apple":1,"banana":2,"guava":3}
for fruit,amount in a.items():
    print(fruit,amount)
```

```python
a = set(('apple', 1, 23))  
print(type(a))
print(a)

for i in a:             # since set is unordered collection we cannot unpack elements as i,j,k
    print(i)
```
