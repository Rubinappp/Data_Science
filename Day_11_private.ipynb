## private :

In python there is no true encapsulation like in c++ ...
Because,if something is private in C++, the compiler enforces it strictly â€” you cannot access it from outside the class.

#### Private in python :: The goal is to avoid accidental access, not make it impossible.So it is not a security feature like in c++,java
Represented by double underscore.__variable

Python uses name mangling to discourage access, not prevent it.

Name mangling means to automatically change ( __variable ) to ( _classname__variable)

```python
class Account:
    def __init__(self,balance=0):
        self.__balance=balance
    def deposit(self,amount):
        self.__balance+=amount
    def displayBalance(self):
        return(self.__balance)
```

```python
ac3=Account()
```

```python
ac3.__balance # error since balance is name mangled so not directly accessed outside
```

```python
# object._className__attributeName
ac3._Account__balance
```

```python
ac3._Account__balance=15000
```

```python
ac3._Account__balance
```

```python
ac3.deposit(200)
```

```python
ac3.deposit(500)
```
