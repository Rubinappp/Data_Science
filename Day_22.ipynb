### Using numpy to generate data 

```python
import matplotlib.pyplot as plt
import numpy as np

# generating data points for a straight line
x = np.arange(0, 11)
y = 2 * x

# print(x, y)

#plotting the straight line
plt.figure(figsize= (60, 5)) # (width,height)
# plot is used to plot lines
plt.plot(x, y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("A Straight Line")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generating data points for a straight line
x = np.arange(0, 11)
y = np.power(x, 2)

# print(x, y)

#plotting the straight line
plt.figure(figsize= (20, 5))
# plot is used to plot lines
plt.plot(x, y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("A Straight Line")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generating data points for a straight line
x = np.arange(0, 11)
y1 = np.power(x, 1)
y2 =np.power(x,2)
y3 =np.power(x,3)
y4 =np.power(x,4)

# print(x, y)

#plotting the straight line
plt.figure(figsize= (20, 5))
# plot is used to plot lines
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("A Straight Line")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generating some data points for plotting a sine plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.figure(figsize=(20, 5))
plt.plot(x, y, label='Sine Wave')
plt.title('Simple Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() # (associated with label during plt.plot) to show box with label
plt.show()
```

### Working with Dataframes

```python
import pandas as pd
# sample dataframe with year and population in billions
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Population': [2.5, 2.7, 2.9, 3.2, 3.5]}
df = pd.DataFrame(data)
df.head()
```

```python
plt.figure(figsize=(15,3))
plt.plot(df['Year'],df['Population'],marker="o",color="r")
plt.title("Population over Years")
plt.xlabel("Year")
plt.ylabel("Population in billions")
plt.show()
```

__Multiple lines on the same plot__

```python
years = np.arange(2000, 2010)
values=np.random.rand(10)
```

```python
values2 = np.random.rand(10)
values3 = np.random.rand(10)
plt.figure(figsize=(4,4))
plt.plot(years, values, label='Series 1')
plt.plot(years, values2, label='Series 2', linestyle='--')
plt.plot(years, values3, label='Series 3', linestyle='-.')
plt.title('Multiple Line Plots')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.show()
```

### Pie Chart

```python
labels=['Apple','Banana','Cherry','Date']
sizes=[215,130,245,210]
#only exploding the first slice
explode=(0,0,0,0.5)
plt.pie(sizes,explode,labels,autopct="%.2f%%",startangle=45)
plt.title('Pie Chart Example')
plt.show()
```

By default, pie charts start at 0, which is the x-axis to the right

startangle rotates the whole pie chart counterclockwise by that many degrees.

##### autopct="%.2f%%"

%.2f  show a floating-point number with 2 decimal places

%%  This is how you actually write a percent symbol

##### explode 

pulls out slices from the center of the pie to highlight them.

Its a list of numbers, same length as sizes.

explode = 0.2
It moves the slice 0.2  radius away from the center.|
