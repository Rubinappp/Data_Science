##  zip function (for equal lengths): 
#### ->built in function (available without importing any module). It creates pairs (tuples)



## zip_longest to the rescue (for unequal lengths):
#### 1) zip_longest is a more versatile version of zip.It also creates pairs (tuples)as zip
#### 2) For the shorter list, it uses a special value (default is None) to fill in the missing spots.


##### using zip function

```python
colors = ["red", "blue", "green"]
sizes = ["small", "medium", "large"]

zipped_pairs = zip(colors, sizes)  # Creates an iterator of tuples
print(zipped_pairs)

for pair in zipped_pairs:
    print(pair)
```

```python
a = [12, 23, 34]
b = [45, 56, 67, 78]
result = []

for i in zip(a, b):
    result.append(i)

print(result)  
```

```python
a = [12, 23, 34]
b = [45, 56, 67, 78]
result = []

for i in zip(a, b):
    result.append(i)
print(result)


if len(a) < len(b):
    for elem in b[len(a):]:
        result.append((None, elem))
        
        
elif len(a) > len(b):
    for elem in a[len(b):]:
        result.append((elem, None))

print(result)
```

```python
a = [12, 23, 34]
b = [45, 56, 67, 78]

# Using list comprehension to create the result
result = [(x, y) for x, y in zip(a, b)]

# Adding remaining elements from the longer list
if len(a) < len(b):
    result.extend([(None, elem) for elem in b[len(a):]])
elif len(a) > len(b):
    result.extend([(elem, None) for elem in a[len(b):]])

print(result)
```
