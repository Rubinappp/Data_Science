# Another way to use subplots without giving multiple  axis while creating subplots

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("sample.csv",index_col=0,parse_dates=True)
```

```python
data.head()
```

```python
figure, axis=plt.subplots(1,2,figsize=(20,6))
axis
```

axes= can be either be single object or an array of axes object 

```python
figure,axis=plt.subplots(3,1,figsize=(20,6),sharex=True)

axis[0].plot(data.index,data["Temperature"],color='red')
axis[0].set_title('Daily Temperature')
axis[0].set_ylabel("Temperature(C)")


axis[1].plot(data.index,data["Humidity"],color="blue")
axis[1].set_title('Daily Humidity')
axis[1].set_ylabel("Humidity(%)")

axis[2].plot(data.index,data['Wind Speed'],color="green")
axis[2].set_title('Daily Wind Speed')
axis[2].set_ylabel('Wind Speed(km/h)')
axis[2].set_xlabel('Date')

plt.show()
```

```python
axis
```

Using for loop

```python
figure,axis=plt.subplots(3,1,figsize=(20,6),sharex=True)
title=["Daily Temperature","Humidity","Wind Speed"]
columns=data.columns
y=["Temperature(C)","Humidity(%)","Wind Speed(km/h)"]
color=["red",'blue','green']
for i in range(0,len(axis)):
        axis[i].plot(data.index,data[columns[i]],c=color[i])
        axis[i].set_title(title[i])
        axis[i].set_ylabel(y[i])
        if(i==len(axis)-1):
            axis[i].set_xlabel("Date")
plt.show()
```

```python
figure,axis=plt.subplots(3,1,figsize=(20,6),sharex=True)
title=["Daily Temperature","Humidity","Wind Speed"]
col=data.columns
y=["Temperature(C)","Humidity(%)","Wind Speed(km/h)"]
color=["red",'blue','green']
i=0
for ax in axis:
        ax.plot(data.index,data[col[i]],c=color[i])
        ax.set_title(title[i])
        ax.set_ylabel(y[i])
        if(i==len(axis)-1):
            ax.set_xlabel("Date")
        i=i+1
plt.show()
```

using loop

```python
fig, axs = plt.subplots(2, 2, sharex=True, figsize = (20, 8))



# Example data
x = np.arange(10)
y1 = [2 * xi for xi in x]
y2 = [xi ** 2 for xi in x]
y3 = [xi ** 3 for xi in x]
y4 = [xi ** 4 for xi in x]
y=[y1,y2,y3,y4]
xlabels=["Linear","Quadratic","Cubic","Quartic"]
k=0
for i in range(0,len(axs)):
    for j in range(0,len(axs)):
        axs[i][j].plot(x,y[k])
        axs[i][j].set_xlabel(xlabels[k])
        k=k+1
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 1)  # 2x2 grid
axs_unflattened = axs.copy()    
# making copy of orginal axs .... axs_unflattened is a copy of that list/array of subplots

for ax in axs_unflattened:
    ax.plot([1,2,3], [1,4,9])
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, figsize = (20, 8))

axs_unflattened = axs.copy() # remember original 2x2 shape
#  [[axs[0,0], axs[0,1]],
#  [axs[1,0], axs[1,1]]]


# Example data
x = np.arange(10)
y1 = [2 * xi for xi in x]
y2 = [xi ** 2 for xi in x]
y3 = [xi ** 3 for xi in x]
y4 = [xi ** 4 for xi in x]

yVals= [y1, y2, y3, y4]
titles = ["Linear", "Quadratic", "Cubic", "Quartic"]

axs = axs.flatten() # make 2*2 grid into 1*4 grid(1D) 
# [axs[0], axs[1], axs[2], axs[3]]  c style order row wise


for ax, y, title in zip(axs , yVals, titles):
    ax.plot(x, y)
    ax.set_title(title)
```

```python

fig, axs = plt.subplots(2, 2, sharex=True, figsize = (20, 8))

axs_unflattened = axs.copy()

# Example data
x = np.arange(10)
y1 = [2 * xi for xi in x]
y2 = [xi ** 2 for xi in x]
y3 = [xi ** 3 for xi in x]
y4 = [xi ** 4 for xi in x]

yVals= [y1, y2, y3, y4]
titles = ["Linear", "Quadratic", "Cubic", "Quartic"]
colors = ['y', 'r', 'g', 'b']

axs = axs.flatten()

for ax, y, title, color in zip(axs , yVals, titles, colors):
    ax.plot(x, y, color = color)
    ax.set_title(title)


```
