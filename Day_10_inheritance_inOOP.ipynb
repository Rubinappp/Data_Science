#### Example2

```python
class Person:
    def __init__(self, name=None, id=None):
        self.name = name
        self.id = id

    def display(self):
        print(self.id)
        print(self.name)

    def details(self):
        print(f"my name is {self.name}")
        print(f"id number is {self.id} ")


class Employee(Person):
    def __init__(self,salary, post):
        super().__init__()  # Calling the __init__ method of the parent class without arguments
        self.salary = salary
        self.post = post

# Example usage
employee1= Employee(50000,"Manager")
print(employee1.name)    # Output: None (not provided)
print(employee1.id)      # Output: None (not provided)
print(employee1.salary)  # Output: 50000
print(employee1.post) 
```

Question: Implement a Course Enrollment System
Create two classes: Student and Course.

1. Student Class
Has the following attributes:

name (string),age (integer),grade (integer from 0 to 100)

Method:----- get_grade() – returns the student's grade.

2. Course Class
Has the following attributes:

name (string),max_students (integer),students (a list to store enrolled Student objects)

Methods:

add_student(student) – adds a student to the course if the limit is not exceeded. Returns False if the course is full.

get_average_grade() – returns the average grade of all enrolled students.

```python
class Student:
    def __init__(self,name,age,grade):
        self.name=name
        self.age=age
        self.grade=grade
    def get_grade(self):
        return self.grade
    
class Course:
    def __init__(self,name,max_students):
        self.name=name
        self.max_students=max_students
        self.students=[]
    def add_student(self,student):
        if(len(self.students)<self.max_students):
            self.students.append(student)
        return False
    def get_average_grade(self):
        value=0
      
        for student in self.students:
            value=value+student.get_grade()
        return value/len(self.students)
        
s1=Student("Rubina",22,96)
s2=Student("Tim",24,95)
s3=Student("Jim",24,80)
course=Course("computer_science",2)
course.add_student(s1)
course.add_student(s2)
print(course.add_student(s3))
course.get_average_grade()
```

#### classmethod
#### Use it when:
- You want to access or modify class variables (shared by all instances)

It is a  method  bound to the class, not the instance.

It can be called using either the class name or an instance, but it always gets the class as its first parameter.


```python
class Person:
    number_of_people=0             # class variable
    def __init__(self,name):
        self.name=name
        Person.add_people()
    @classmethod
    def add_people(cls):
        cls.number_of_people+=1
    @classmethod
    def no_of_people(cls):
        return cls.number_of_people
p1=Person("Rubina")
p2=Person("hello")
Person.no_of_people()
```

#### static method
What is a @staticmethod?

A static method is a method that:

- Belongs to the class, not the instance.

- Doesn’t take self or cls as its first argument.

- Can be called using either the class name or an object, but:

            - It doesn’t know anything about the class or instance.

            - It can’t access or modify class or instance variables.

```python

class algebra:
    @staticmethod
    def add10(x):
        return x+10
    @staticmethod
    def add1(x):
        return x+1


algebra.add10(5)
```
