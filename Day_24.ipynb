
## Practice Question: Analyzing Daily Sales and Advertising Spend Using Subplots
### Objective: Create a visual analysis of daily sales and advertising spend over one year. Your task is to create visualisations using subplots in Matplotlib. This will help in understanding how variations in advertising spend could potentially affect sales.

### Background:
- Assume you're working with a retail company monitoring their daily performance. The dataset will include:
1. **Sales**: Daily sales in USD, which fluctuates based on multiple factors including advertising.
2. **Ad_Spend**: Daily advertising spending in USD, which the company adjusts based on their monthly strategy and market response



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sales_advertising.csv",index_col=0,parse_dates=True)
```

It tells Pandas to automatically detect and convert date-like columns into actual datetime objects instead of keeping them as plain strings.

```python
df.head()
```

```python
df.index
```

```python
date=list(df.index)
date
```

```python
sales=list(df["Sales"])
```

```python
Ad_Spend=list(df["Ad_Spend"])
```


## Tasks:
### Data Visualization:
1. Create a figure with two subplots stacked vertically.
2. Plot the Sales data in the first subplot as a line plot.
3. Plot the Ad_Spend in the second subplot as a bar plot to emphasize daily changes.
4. Ensure each plot is properly labeled with titles, x-labels, y-labels, and legends.

```python
figure,(ax1,ax2)=plt.subplots(2,1,figsize=(20,5))

ax1.plot(date,sales,label="Daily Sales")
ax1.set_title("Daily Sales over one year")
ax1.set_xlabel("Dates")
ax1.set_ylabel("Sales(USD)")
ax1.legend()

ax2.bar(date,Ad_Spend,label="Ad_Spend")
ax2.set_title("Daily Advertising spend over one year")
ax2.set_xlabel("Dates")
ax2.set_ylabel("Ad_Spend(USD)")

figure.tight_layout(pad=7) # manage spacing by using this or by simply changing figsize

plt.show()
```

```python
figure,(ax1,ax2)=plt.subplots(2,1,figsize=(20,5))

ax1.plot(df.index,df["Sales"],label="Daily Sales")
ax1.set_title("Daily Sales over one year")
ax1.set_xlabel("Dates")
ax1.set_ylabel("Sales(USD)")
ax1.legend()

ax2.bar(df.index,df["Ad_Spend"],label="Ad_Spend")
ax2.set_title("Daily Advertising spend over one year")
ax2.set_xlabel("Dates")
ax2.set_ylabel("Ad_Spend(USD)")

figure.tight_layout(pad=5) # manage spacing by using this or by simply changing figsize

plt.show()
```
