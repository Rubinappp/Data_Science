##### write a function that gives lower sentence as well as the upper sentence

```python
def convert_sentence(sentence,a):
    if a:
        return sentence.upper()
    else:
        return sentence.lower()
    
sentence="The sky is too high"
a=bool(int(input("Enter 0 or 1 :")))
convert_sentence(sentence,a)
```

##### write a function to greet using default value in argument

```python
def greet(name,greet="hi"):
    return(f"{greet}{name}")
    
greeting=greet("rubina","welcome")
print(greeting)
print(greet("rubina"))
```

##### write a function to return odd and even numbers

 type hint:
 
 it's optional in Python.Improve readability and documentation
 
 paramater : list /bool/str/int/dict....



```python
def odd_even(num : list, reteven : bool = True ):
    list1=[]
    list2=[]
    if reteven:
        for i in num:
            if(i%2==0):
                list1.append(i)
        return list1
    else:
        for i in num:
            if(i%2!=0):
                list2.append(i)
        return list2
num=[1,2,3,4,5,6,7]
result=odd_even(num)
print(result)
result1=odd_even(num,reteven=False)
print(result1)
```
